```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from 0 to 2 with small step for smoothness
x = np.linspace(0, 2 * np.pi, 1000)  # 1000 points for smooth curve
y = np.sin(x)  # Sine function

# Create the plot
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='sin(x)', color='blue', linewidth=2)

# Add labels and title
plt.title("Smooth Sine Wave", fontsize=14)
plt.xlabel("x (radians)")
plt.ylabel("sin(x)")

# Add grid and legend
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Show the plot
plt.show()
```
