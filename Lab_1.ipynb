```python
#Recall Nyquist Limit
#Bandwidth cannot be more than Nyquist limit
#Bandwidth is found by fft on data, square the abs val

#To harvest bandwidth data

sample_rate = 2.2e6
bandwidth = 1/sample_rate


Nylim = sample_rate/2

if bandwidth > Nylim:
    print(f"Bandwidth is too high by:  {Nylim-bandwidth}")
else:
    print(f"Bandwidth is within limit")
```
